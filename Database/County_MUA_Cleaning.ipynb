```python
import pandas as pd
mr = pd.read_csv('Resources/death_all.csv')
ir = pd.read_csv('Resources/incd_all.csv')
mua = pd.read_csv('Resources/MUA_index_cook.csv')
state_abbreviation = pd.read_csv('Resources/state_abbreviation_cook.csv')
```

# Mortality = mr

```python
mr.head()
```

```python
list(mr)
```

```python
mr.drop([ 'Unnamed: 1',
 'Unnamed: 2',
 'Unnamed: 4',
 'Unnamed: 5',
 'Unnamed: 6',
 'Unnamed: 7',
 'Unnamed: 8',
 'Unnamed: 9',
 'Unnamed: 10',
 'Unnamed: 11',
 'Unnamed: 12',
 'Unnamed: 13'], 1, inplace=True)
mr = mr.drop([1,2,3,4,5,6,7,8], axis=0)
mr = mr.reset_index(drop=True)
mr.head()
```

```python
mr = mr.rename(columns={"Death Rate Report for United States by County":"County_State", "Unnamed: 3":"Age-Adjusted Death Rate"})
mr.head()
```

```python
mr.dropna()
mr.head()
```

```python
mr['Age-Adjusted Death Rate'] = mr['Age-Adjusted Death Rate'].str.replace(r'\(.*\)', '')
print(mr.shape)
mr.head()
```

```python
mr['Exract'] = mr['County_State'].str.extract(r'(County|Parish).*')
mr
```

```python
mr = mr[mr['Exract'].notna()]
mr.head()
```

```python
mr.drop(['Exract'], 1, inplace=True)
mr = mr.reset_index(drop=True)
mr.head()
```

```python
mr2 = mr["County_State"].str.split(",", n = 1, expand = True)
mr["County"]= mr2[0]
mr["Primary_State_Name"]= mr2[1]
mr.drop(columns =["County_State"], inplace = True)
mr.head()
```

# Incidence Rate = ir

```python
ir.head()
```

```python
list(ir)
```

```python
ir.drop([ 'Unnamed: 1',
 'Unnamed: 2',
 'Unnamed: 4',
 'Unnamed: 5',
 'Unnamed: 6',
 'Unnamed: 7',
 'Unnamed: 8',
 'Unnamed: 9',
 'Unnamed: 10',
 'Unnamed: 11',
 'Unnamed: 12',
 'Unnamed: 13'], 1, inplace=True)
ir = ir.drop([1,2,3,4,5,6,7,8], axis=0)
ir = ir.reset_index(drop=True)
ir.head()
```

```python
ir = ir.rename(columns={"Incidence Rate Report for United States by County":"County_State", "Unnamed: 3":"Age-Adjusted Incidence Rate"})
ir.head()
```

```python
ir.dropna()
ir.head()
```

```python
ir['Age-Adjusted Incidence Rate'] = ir['Age-Adjusted Incidence Rate'].str.replace(r'\(.*\)', '')
print(ir.shape)
ir.head()
```

```python
ir['Exract'] = ir['County_State'].str.extract(r'(County|Parish).*')
ir
```

```python
ir = ir[ir['Exract'].notna()]
ir.head()
```

```python
ir.drop(['Exract'], 1, inplace=True)
ir = ir.reset_index(drop=True)
ir.head()
```

```python
ir2 = ir["County_State"].str.split(",", n = 1, expand = True)
ir["County"]= ir2[0]
ir["Primary_State_Name"]= ir2[1]
ir.drop(columns =["County_State"], inplace = True)
ir.head()
```

```python
ir['Primary_State_Name'] = ir['Primary_State_Name'].str.replace(r'\(.*\)', '')
print(ir.shape)
ir.head()
```

# Merge Mortality, Incidence, MUA Index, and State Abbreviation

```python
new_df = pd.merge(mr, ir)
new_df
```

```python
column_names = ["Age-Adjusted Death Rate", "Age-Adjusted Incidence Rate", "County", "Primary_State_Name"]
new_df = new_df.reindex(columns=column_names)
new_df.head()
```

```python
new_df['Primary_State_Name'] = new_df['Primary_State_Name'].str.replace('\d+', '', regex=True)
new_df.head()
```

```python
new_df.to_csv('Resources/mort_incd.csv', index=False)
```
