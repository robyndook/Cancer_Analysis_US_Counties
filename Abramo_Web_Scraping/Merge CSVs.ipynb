```python
import pandas as pd
```

```python
race_inc = pd.read_csv('race_inc.csv')
sex_inc = pd.read_csv('sex_inc.csv')
age_inc = pd.read_csv('age_inc.csv')
```

```python
race_inc = race_inc.drop('Unnamed: 0', axis='columns')
sex_inc = sex_inc.drop('Unnamed: 0', axis='columns')
age_inc = age_inc.drop('Unnamed: 0', axis='columns')
```

```python
race_inc 
```

```python
sex_inc
```

```python
age_inc
```

```python
groupi = [race_inc, sex_inc,age_inc]
```

```python
Incidence_df = pd.concat(groupi)
```

```python
Incidence_df
```

```python
Incidence_df[Incidence_df["County"].str.contains("Washington County")].sort_values(["State", "County"])
```

```python
Incidence_df.to_csv('Incidence_data.csv', index=False)
```
